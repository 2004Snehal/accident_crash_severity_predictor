<h1>Feature Engineering for Categoricals

```python
#converting all false to 0 and true to 1
df['Alcohol_Consumption_Yes']=df['Alcohol_Consumption_Yes'].astype(int)
df['Seatbelt_Usage_Yes']=df['Seatbelt_Usage_Yes'].astype(int)
df['Road_Surface_Condition_Icy']=df['Road_Surface_Condition_Icy'].astype(int)
df['Road_Surface_Condition_Wet']=df['Road_Surface_Condition_Wet'].astype(int)
df['Vehicle_Type_Heavy Vehicle']=df['Vehicle_Type_Heavy Vehicle'].astype(int)
df['Vehicle_Type_T.W']=df['Vehicle_Type_T.W'].astype(int)
df['Crash_Type_Rear-end']=df['Crash_Type_Rear-end'].astype(int)
df.head()
```

```python
df['Safety_impact']=df['Seatbelt_Usage_Yes']+ ((df['Alcohol_Consumption_Yes']+1)%2)
df.drop(['Seatbelt_Usage_Yes','Alcohol_Consumption_Yes'],axis=1,inplace=True)
```

```python
# df['Crash_Heavy']=df['Vehicle_Type_Heavy Vehicle']+df['Crash_Type_Rear-end']
# df['Crash_TW']=df['Vehicle_Type_T.W']+df['Crash_Type_Rear-end']
# df.drop(['Vehicle_Type_Heavy Vehicle','Vehicle_Type_T.W','Crash_Type_Rear-end'],axis=1,inplace=True)
```

```python
df.drop(['Gender_Male'],axis=1,inplace=True)
df.drop(['Crash_Type_Rear-end'], axis=1,inplace=True)
```

```python
df['road_urban_conditionIcy']=((df['Road_Surface_Condition_Icy']+1)%2)+df['Road_Type_Urban']
df['road_urban_conditionWet']=((df['Road_Surface_Condition_Wet']+1)%2)+df['Road_Type_Urban']
df.drop(['Road_Surface_Condition_Icy','Road_Surface_Condition_Wet','Road_Type_Urban'],axis=1,inplace=True)
```

```python
X=df.drop(['Crash_Severity'],axis=1)
y=df['Crash_Severity']
X.head()
```

```python
#for Y convert to labels
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Minor injury','Major injury','Fatal crash']])
y=oe.fit_transform(y.values.reshape(-1,1))
```

```python
print(y)
```

```python
#Spliting into test train
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=1)
```

<h2>Models

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import xgboost as xgb
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
# Importing DecisionTreeClassifier, AdaBoostClassifier, and BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier
```

```python
# Dictionary of models
models = {
    'Logistic Regression': LogisticRegression(max_iter=10000),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'XGBoost': xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC(),
    'Naive Bayes': GaussianNB(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'Bagging': BaggingClassifier(random_state=42)

}

# Training and evaluating each model
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Predictions
    y_pred = model.predict(X_test)

    # Evaluation
    print(f"\n{model_name} Results:")
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")


    # Display plot
    plt.show()
```
